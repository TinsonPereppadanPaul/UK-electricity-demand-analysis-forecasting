```python
import pandas as pd
import matplotlib
import numpy as np
import xgboost as xgb



df['settlement_date'] = pd.to_datetime(df['settlement_date'])
df.set_index('settlement_date', inplace=True)

last_date = df.index[-1]
future_dates = pd.date_range(start=last_date + pd.Timedelta(minutes=30), periods=1488, freq='30min')
future = pd.DataFrame(index=future_dates)
future.index.name = 'settlement_date'

future['period_hour'] = future.index.hour
future['day_of_week'] = future.index.dayofweek
future['month'] = future.index.month

future['is_holiday'] = 0  

demand_history = df['england_wales_demand'].tolist()[-48:]  


df['period_hour'] = df.index.hour
df['day_of_week'] = df.index.dayofweek
df['month'] = df.index.month
df['is_holiday'] = 0  
df['demand_lag_1'] = df['england_wales_demand'].shift(1)
df['demand_lag_48'] = df['england_wales_demand'].shift(48)
df.dropna(inplace=True)

features = ['period_hour','day_of_week','month','is_holiday','demand_lag_1','demand_lag_48']
X_train = df[features]
y_train = df['england_wales_demand']

model = xgb.XGBRegressor(n_estimators=500, learning_rate=0.05, max_depth=6)
model.fit(X_train, y_train)

predictions = []

for i in range(len(future)):
    row = future.iloc[i].copy()
    row['demand_lag_1'] = demand_history[-1]
    row['demand_lag_48'] = demand_history[-48]

    pred = model.predict(row[features].values.reshape(1, -1))[0]
    predictions.append(pred)

    demand_history.append(pred)

future['predicted_demand'] = predictions

print(future.head())
```
